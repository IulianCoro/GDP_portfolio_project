```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('all_data.csv')
```

```python
print(df.head())
```

PROJECT SCOPING
- clean data and check for missing values
- summary statistics for each country
- correlation life expectancy and gdp
- hypothesis testing
- is there any relation between GDP and life expectancy
- the relation is the same for all countries?


```python
df.info()
```

```python
df = df.rename(columns={'Country':'country', 'Year':'year', 'Life expectancy at birth (years)':'life', 'GDP':'gdp'})
```

```python
plt.figure(figsize=(15,12))

sns.displot(df.gdp, rug = True, rug_kws={'color':'black'}, alpha=0.4)
plt.xlabel("GDP in Trillions of U.S. Dollars");
plt.ylabel('Count')
plt.title('Distribution of GDP values', fontsize=14, fontweight='bold')

gdp_mean = np.mean(df.gdp)
gdp_median = np.median(df.gdp)
plt.axvline(x = gdp_mean, color='blue', label = 'Mean', linestyle='dotted')
plt.text(gdp_mean*1.1, 30, 'Mean = {:.2e}'.format(gdp_mean), rotation=90, fontweight='bold')
plt.axvline(x = gdp_median, color='blue', label = 'Median', linestyle='dashdot')
plt.text(gdp_median*1.5, 30, 'Median = {:.2e}'.format(gdp_median), rotation=90, fontweight='bold')

plt.show()
plt.clf()
```

```python
plt.figure(figsize=(15,12))

sns.displot(df.life, rug = True, rug_kws={'color':'black'}, alpha=0.4)
plt.xlabel("Life expectancy at bith in years");
plt.ylabel('Count')
plt.title('Distribution life expectancy values', fontsize=14, fontweight='bold')

life_exp_mean = np.mean(df.life)
life_exp_median = np.median(df.life)
plt.axvline(x = life_exp_mean, color='blue', label = 'Mean', linestyle='dotted')
plt.text(life_exp_mean*0.98, 15, 'Mean = {:.2f}'.format(life_exp_mean), rotation=90, fontweight='bold')
plt.axvline(x = life_exp_median, color='blue', label = 'Median', linestyle='dashdot')
plt.text(life_exp_median*0.985, 15, 'Median = {:.2f}'.format(life_exp_median), rotation=90, fontweight='bold')

plt.show()
plt.clf()
```

```python
df_means = df.drop("year", axis = 1).groupby("country").mean().reset_index()
df_medians = df.drop("year", axis = 1).groupby('country').median().reset_index()

df_means
df_medians
```

```python
plt.figure(figsize=(12,9))

sns.barplot(x='country', y='gdp', data=df_means)
plt.xlabel('Country')
plt.ylabel('GDP mean')
plt.title('GDP mean for various countries', fontweight='bold')
```

```python
plt.figure(figsize=(12,9))

sns.barplot(x='country', y='life', data=df_means)
plt.xlabel('Country')
plt.ylabel('Life expectancy mean')
plt.title('Life expectancy mean for various countries', fontweight='bold')
```

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
axes[0] = sns.violinplot(ax=axes[0], x=df.gdp, y=df.country)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
axes[0].set_ylabel('Country')
axes[1] = sns.violinplot(ax=axes[1], x=df.life, y=df.country)
axes[1].set_xlabel("Life expectancy at birth (years)");
axes[1].set_ylabel('Country')
```

```python
life_exp_chile = df.life[df.country == 'Chile']
life_exp_china = df.life[df.country == 'China']
life_exp_germany = df.life[df.country == 'Germany']
life_exp_mexico = df.life[df.country == 'Mexico']
life_exp_us = df.life[df.country == 'United States of America']
life_exp_zimbabwe = df.life[df.country == 'Zimbabwe']

gdp_chile = df.gdp[df.country == 'Chile']
gdp_china = df.gdp[df.country == 'China']
gdp_germany = df.gdp[df.country == 'Germany']
gdp_mexico = df.gdp[df.country == 'Mexico']
gdp_us = df.gdp[df.country == 'United States of America']
gdp_zimbabwe = df.gdp[df.country == 'Zimbabwe']

corr_chile = pearsonr(life_exp_chile, gdp_chile)[0]
corr_china = pearsonr(life_exp_china, gdp_china)[0]
corr_germany = pearsonr(life_exp_germany, gdp_germany)[0]
corr_mexico = pearsonr(life_exp_mexico, gdp_mexico)[0]
corr_us = pearsonr(life_exp_us, gdp_us)[0]
corr_zimbabwe = pearsonr(life_exp_zimbabwe, gdp_zimbabwe)[0]

fig1 = plt.figure(figsize=(15,12))

plt.subplot(2, 3, 1)
plt.scatter(x=life_exp_chile, y=gdp_chile, color='red')
plt.title('Chile \n Corr = %f' %corr_chile)
plt.xlabel('Life expectancy at birth [years]')
plt.ylabel('GDP')
plt.grid()

plt.subplot(2, 3, 2)
plt.scatter(x=life_exp_china, y=gdp_china, color='orange')
plt.title('China \n Corr = %f' %corr_china)
plt.xlabel('Life expectancy at birth [years]')
plt.ylabel('GDP')
plt.grid()

plt.subplot(2, 3, 3)
plt.scatter(x=life_exp_germany, y=gdp_germany, color='green')
plt.title('Germany \n Corr = %f' %corr_germany)
plt.xlabel('Life expectancy at birth [years]')
plt.ylabel('GDP')
plt.grid()

plt.subplot(2, 3, 4)
plt.scatter(x=life_exp_mexico, y=gdp_mexico, color='blue')
plt.title('Mexico \n Corr = %f' %corr_mexico)
plt.xlabel('Life expectancy at birth [years]')
plt.ylabel('GDP')
plt.grid()

plt.subplot(2, 3, 5)
plt.scatter(x=life_exp_us, y=gdp_us, color='purple')
plt.title('USA \n Corr = %f' %corr_us)
plt.xlabel('Life expectancy at birth [years]')
plt.ylabel('GDP')
plt.grid()

plt.subplot(2, 3, 6)
plt.scatter(x=life_exp_zimbabwe, y=gdp_zimbabwe, color='black')
plt.title('Zimbabwe \n Corr = %f' %corr_zimbabwe)
plt.xlabel('Life expectancy at birth [years]')
plt.ylabel('GDP')
plt.grid()

plt.suptitle('Life expectancy VS. GDP for various countries', fontsize=14, fontweight='bold')
plt.subplots_adjust(hspace=0.3, wspace=0.3)
plt.show()
plt.clf()
```

```python
fig2 = plt.figure(figsize=(15,12))

years = df.year.unique()

ax1 = plt.subplot(1, 2, 1)
plt.plot(years, gdp_chile, marker='s', color='red', label='Chile')
plt.plot(years, gdp_china, marker='s', color='orange', label='China')
plt.plot(years, gdp_germany, marker='s', color='green', label='Germany')
plt.plot(years, gdp_mexico, marker='s', color='blue', label='Mexico')
plt.plot(years, gdp_us, marker='s', color='purple', label='USA')
plt.plot(years, gdp_zimbabwe, marker='s', color='black', label='Zimbabwe')
plt.legend()
plt.grid()
plt.xlabel('Year')
plt.ylabel('GDP')
ax1.set_xticks(years)
ax1.set_xticklabels(years, rotation=45)
plt.title('GDP', fontsize=14)

ax2 = plt.subplot(1, 2, 2)
plt.plot(years, life_exp_chile, marker='s', color='red', label='Chile')
plt.plot(years, life_exp_china, marker='s', color='orange', label='China')
plt.plot(years, life_exp_germany, marker='s', color='green', label='Germany')
plt.plot(years, life_exp_mexico, marker='s', color='blue', label='Mexico')
plt.plot(years, life_exp_us, marker='s', color='purple', label='USA')
plt.plot(years, life_exp_zimbabwe, marker='s', color='black', label='Zimbabwe')
plt.legend()
plt.grid()
plt.xlabel('Year')
plt.ylabel('GDP')
ax2.set_xticks(years)
ax2.set_xticklabels(years, rotation=45)
plt.title('Life expectancy at birth in years', fontsize=14)

plt.suptitle('Evolution of GDP and life expectancy in time for various countries', fontsize=14, fontweight='bold')
plt.subplots_adjust(hspace=0.3, wspace=0.3)
plt.show()
plt.clf()
```
